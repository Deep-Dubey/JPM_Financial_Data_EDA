#Random Forest

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.ensemble import RandomForestClassifier
```

```python
flag = pd.get_dummies(jp,drop_first=True)
flag
```

```python
X = jp.drop(['default_flag','transaction_amount'],axis=1)
y = jp['default_flag']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify=y,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print("Before SMOTE (Training data):")
print(y_train.value_counts())
print("\nAfter SMOTE (Resampled Training data):")
print(pd.Series(y_train_res).value_counts())
```

```python
rf = RandomForestClassifier(n_estimators=100,class_weight='balanced', random_state=42)
```

```python
cv_params = {
    'max_depth':[2,3,4,5,None],
    'min_samples_leaf':[1,2,3],
    'min_samples_split':[2,3,4],
    'max_features':[2,3,4],
    'n_estimators':[75,100,125,150]
    }
```

```python
scoring = {"accuracy":"accuracy","precision":"precision","recall":"recall","f1":"f1"}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_cv = RandomizedSearchCV(
    rf,
    cv_params,
    n_iter=20,
    scoring=scoring,
    refit="f1",
    cv=3,
    n_jobs=-1,
    verbose=2
)
```

```python
rf_cv.fit(X_train_res, y_train_res)
```

```python
y_pred = rf_cv.best_estimator_.predict(X_test)
```

```python
#Probability
y_prob = rf_cv.best_estimator_.predict_proba(X_test)[:,1]
y_prob
```

```python
rf_cv.best_score_
```

```python
rf_cv.best_params_
```

```python
print("Accuracy",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=np.unique(y_test))
disp.plot()
```

```python
print(f"ROC AUC Score: {roc_auc_score(y_test, y_prob)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get feature importances from the best estimator
feature_importances = rf_cv.best_estimator_.feature_importances_

# Create a DataFrame for better visualization
features_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

# Sort the features by importance in descending order
features_df = features_df.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=features_df)
plt.title('Feature Importances from Random Forest Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

### What the Current Model Metrics Tell Us:

*   **Accuracy:** The proportion of correctly classified instances. A high accuracy (like 1.0) means the model is performing very well on the test data.
*   **Precision:** For class 1 (default), it's the proportion of actual defaults among those predicted as default. For class 0 (no default), it's the proportion of actual non-defaults among those predicted as non-default.
*   **Recall:** For class 1 (default), it's the proportion of actual defaults that were correctly identified. For class 0, it's the proportion of actual non-defaults that were correctly identified.
*   **F1-Score:** The harmonic mean of precision and recall. It's a good metric for imbalanced datasets like this one, where default cases are much fewer.
*   **Confusion Matrix:** Shows the number of true positives, true negatives, false positives, and false negatives. In our case, `[[257, 0], [0, 36]]` indicates:
    *   257 True Negatives (correctly predicted no default)
    *   0 False Positives (incorrectly predicted default when it was no default)
    *   0 False Negatives (incorrectly predicted no default when it was default)
    *   36 True Positives (correctly predicted default)
*   **ROC AUC Score:** Measures the model's ability to distinguish between classes across all possible classification thresholds. An AUC of 1.0 indicates a perfect classifier.
